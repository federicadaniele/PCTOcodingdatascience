```python
# PCTO in Coding & Data Science/Modulo 1: Introduzione al coding

# Importare libraries/packages/modules:
import pandas as pd
from numpy import array
```

```python
## Le variabili: stringhe, numeri, liste e dizionari

# Creazione di stringhe e numeri
nome    = "Federica"        # una stringa
eta     = 31                # un numero

print(nome)                  # cosa abbiamo creato?
print(eta)
```

```python
# Creazione di una lista
mylist = [ 'Federica', 'Elena']  
print(mylist)
print(mylist[0])            # esempio di indicizzazione
```

```python
# Modificare una lista
mylist.append('Giulia')     # ['listname'].append è un metodo
print(mylist)
```

```python
# Un tipo speciale di lista: le arrays, su cui si possono svolgere operazioni matematiche
myarray = array([31,30])
myarray/2
```

```python
mynewlist = [31,30]
mynewlist/2
```

```python
mydict = {
    'Federica': 31,
    'Elena': 30}
print(mydict)
```

```python
# Modificare un dizionario
mydict['Giulia'] = 28
print(mydict)
```

```python
# Dizionario e liste si possono combinare!
mynewdict = {
    'nome': ['Federica','Elena','Giulia'],
    'eta': [31,30,28]}      # una chiave per ciascun pezzo di informazione relativo a ciascun individuo (per es.: una per il nome e una per l'età)
print(mynewdict)
```

```python
# l'ordine non è casuale: la struttura delle nostre variabili riflette le connessioni tra i vari pezzi di informazione
pd.DataFrame.from_dict(mynewdict)

# Quali e quante sono le colonne? Da cosa sono identificate?
# Quali e quante sono le righe? Da cosa sono identificate?
```
